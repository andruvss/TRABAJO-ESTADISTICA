Análisis descriptivo <ITEM I>

Según la base de datos entregada, podemos determinar que la población de estudio se trata de estudiantes matriculados que esten cursando alguna carrera en las regiones de Antofagasta, Atacama, La Araucania y Los Rios.

<Variables>

Edad: Es una variable CUANTITATIVA CONTINUA ya que se expresa como multiples numeros en un rango de valores que se miden en una escala continua del tiempo.

Nombre carrera: Es una variable CUALITATIVA NOMINAL ya que se tratan de etiquetas que no poseen una jerarquia natural no hay carreras mayores o menores.

Tipo de Institución: Se trata de una variable CUALITATIVA ORDINAL ya que existe una jerarquía academica entre los institutos o centros de formacion con respecto a las universidades.

Area Conocimiento: Se trata de una variable CUALITATIVA NOMINAL, esto se debe a que en un instituto no hay una carrera que sea mas importante o se encuentre sobre otra, no vemos una jerarquía.

Valor Arancel(Pesos): Se trata nuevamente de una variable CUANTITATIVA CONTINUA, ya que vemos claramente un rango de valores, se repiten pero no son todos identicos.

Región Sede: Es una variable CUALITATIVA NOMINAL ya que representa una ubicación geográfica y que no esta por sobre ninguna otra de las regiones.

Jornada: Esta variable CUALITATIVA NOMINAL indica en que turno es que asiste a las clases que le imparten, nuevamente y naturalmente no hay jornadas mayores que otras.

Nombre de institución: Otra variable CUALITATIVA NOMINAL que representa el nombre de una institución, estas por si solas NO estan por encima de otra.

Nivel de estudio: Es una variable CUALITATIVA ORDINAL que indica el programa de estudio que etsa cursando el alumno, en este caso si hay niveles de estudio con más relevancia que otros, como lo es el caso de un posgrado o un técnico.

Desarrollo de las preguntas <ITEM II>

°1 ¿Hay áreas del conocimiento donde las carreras sean más caras?.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Leer el archivo excel
df = pd.read_excel('BLOQUE 4.xlsx')

# =====================================================================
# Interpretación de medidas de tendencia central.
# =====================================================================

media_por_area = df.groupby('AREA CONOCIMIENTO')[ 'VALOR ARANCEL (PESOS)'].mean()

mediana_por_area = df.groupby('AREA CONOCIMIENTO')[ 'VALOR ARANCEL (PESOS)'].median()

moda_por_area = df.groupby('AREA CONOCIMIENTO')['VALOR ARANCEL (PESOS)'].agg(
    lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan
)
print(f'Promedio de arancel por cada area de conocimiento es: {media_por_area}\nLa Mediana: {mediana_por_area}\n La Moda: {moda_por_area}')

# =====================================================================
# Interpretación de medidas de dispercion.
# =====================================================================

#Calculamos el rango de ambas variables.
maximo_arancel_por_area = df.groupby('AREA CONOCIMIENTO')[ 'VALOR ARANCEL (PESOS)'].max()
minimo_arancel_por_area = df.groupby('AREA CONOCIMIENTO')[ 'VALOR ARANCEL (PESOS)'].min()
rango_ar = maximo_arancel_por_area - minimo_arancel_por_area
print(f'Rango: {rango_ar} Arancel.')

#Desviación estándar
desviacion_valor_x_area_c = df.groupby('AREA CONOCIMIENTO')[ 'VALOR ARANCEL (PESOS)'].std()

#Coeficiente de variación
cv_arancel = desviacion_valor_x_area_c / media_por_area * 100

print(f'El rango de los valores de arancel ')


# =====================================================================
# Analisis de la tabla de frecuencia
# =====================================================================
tabla_2 = df[['AREA CONOCIMIENTO', 'VALOR ARANCEL (PESOS)']]

# =====================================================================
# Calculamos estadísticas por área de conocimiento
# =====================================================================
resumen_arancel = tabla_2.groupby('AREA CONOCIMIENTO')['VALOR ARANCEL (PESOS)'].agg(
    Frecuencia_absoluta = 'count',     # número de carreras en esa área
    Promedio = 'mean',                 # costo promedio
    Mediana = 'median',                # valor central
    Máximo = 'max',                    # arancel más alto
    Mínimo = 'min'                     # arancel más bajo
)

# =====================================================================
# Redondeamos a enteros para que se vea más limpio
# =====================================================================
resumen_arancel = resumen_arancel.round(0)

# =====================================================================
# Ordenamos de mayor a menor según el Promedio
# =====================================================================
resumen_arancel = resumen_arancel.sort_values('Promedio', ascending=False)

# =====================================================================
# Mostramos la tabla
# =====================================================================
print(resumen_arancel.head(10))  # mostramos solo las 10 áreas más caras

# =====================================================================
# Preparamos las variables para el gráfico
# =====================================================================
categorias_area = resumen_arancel.index                # nombres de las áreas
valores_promedio = resumen_arancel['Promedio'].values              # costos promedios

# =====================================================================
# Creamos el gráfico circular
# =====================================================================
fig, ax = plt.subplots(figsize=(8,6))
circular_arancel = ax.pie(
    valores_promedio, 
    labels=categorias_area, 
    autopct='%.1f%%', 
    startangle=90
)

# =====================================================================
# Personalización
# =====================================================================
ax.set_title("Distribución del costo promedio de arancel por área de conocimiento")

plt.show()

# =====================================================================
# Relacion con otras variables
# =====================================================================

# Promedio de arancel por área y duración
promedio_area_duracion = df.groupby(['AREA CONOCIMIENTO', 'DURACION TOTAL CARRERA (SEMESTRES)'])['VALOR ARANCEL (PESOS)'].mean().reset_index()

# Redondeamos
promedio_area_duracion['VALOR ARANCEL (PESOS)'] = promedio_area_duracion['VALOR ARANCEL (PESOS)'].round(0)

print("\nArancel promedio por área de conocimiento y duración de la carrera:")
print(promedio_area_duracion.sort_values(['VALOR ARANCEL (PESOS)'], ascending=False).head(10))

# Promedio de arancel por área y provincia
promedio_area_provincia = df.groupby(['AREA CONOCIMIENTO', 'PROVINCIA SEDE'])['VALOR ARANCEL (PESOS)'].mean().reset_index()

# Redondeamos
promedio_area_provincia['VALOR ARANCEL (PESOS)'] = promedio_area_provincia['VALOR ARANCEL (PESOS)'].round(0)

print("\nArancel promedio por área de conocimiento y provincia:")
print(promedio_area_provincia.sort_values(['VALOR ARANCEL (PESOS)'], ascending=False).head(10))
```

2° ¿Qué influencia tiene la edad en el tipo de institución a la que ingresan los estudiantes?

```python
# =====================================================================
# Interpretación de medidas de tendencia central.
# =====================================================================

media_por_institucion = df.groupby('TIPO DE INSTITUCION')['EDAD'].mean()
mediana_por_institucion = df.groupby('TIPO DE INSTITUCION')['EDAD'].median()

# Corrección del uso de .mode() con groupby usando agg + lambda
moda_por_institucion = df.groupby('TIPO DE INSTITUCION')['EDAD'].agg(
    lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan
)

print("Promedio de edad por tipo de institución:\n", media_por_institucion)
print("\nMediana de edad por tipo de institución:\n", mediana_por_institucion)
print("\nModa de edad por tipo de institución:\n", moda_por_institucion)

# =====================================================================
# Interpretación de medidas de dispercion.
# =====================================================================

maximo_edad_por_institucion = df.groupby('TIPO DE INSTITUCION')['EDAD'].max()
minimo_edad_por_institucion = df.groupby('TIPO DE INSTITUCION')['EDAD'].min()
rango_edad = maximo_edad_por_institucion - minimo_edad_por_institucion
print("\nRango de edad por tipo de institución:\n", rango_edad)

desviacion_edad_por_institucion = df.groupby('TIPO DE INSTITUCION')['EDAD'].std()
print("\nDesviación estándar de la edad por tipo de institución:\n", desviacion_edad_por_institucion)

# Coeficiente de variación (en %)
cv_edad = desviacion_edad_por_institucion / media_por_institucion * 100
print("\nCoeficiente de variación de la edad por tipo de institución (%):\n", cv_edad)

# =====================================================================
# Creamos una tabla de frecuencias entre Edad y Tipo de Institución
# =====================================================================

# Contamos cuántos estudiantes hay en cada combinación de Edad e Institución
cuenta_edad_institucion = df.groupby(['EDAD', 'TIPO DE INSTITUCION']).size()

# =====================================================================
# Construimos un DataFrame con las medidas de frecuencia
# =====================================================================

tabla_edad_institucion = pd.DataFrame({
    'Frecuencia absoluta': cuenta_edad_institucion,
    'Frecuencia absoluta acumulada': cuenta_edad_institucion.cumsum(),
    'Frecuencia relativa (%)': round(cuenta_edad_institucion / len(df) * 100, 2),
    'Frecuencia relativa acumulada (%)': round((cuenta_edad_institucion / len(df) * 100).cumsum(), 2)
})

# =====================================================================
# Ordenamos por la Frecuencia absoluta (de mayor a menor)
# =====================================================================
tabla_edad_institucion = tabla_edad_institucion.sort_values('Frecuencia absoluta', ascending=False)

# =====================================================================
# Mostramos las primeras 15 filas para no saturar la salida
# =====================================================================
print(tabla_edad_institucion.head(15))

# ANALISIS DE TABLA DE FRECUENCIA O INTERPRETACION

# GRAFICO PARA LA INFLUENCIA DE EDAD EN EL TIPO DE INSTITUCION DE INGRESO
tabla = pd.crosstab(
    df["RANGO EDAD"], 
    df["TIPO DE INSTITUCION"], 
    normalize="index"
) * 100   # porcentaje por rango de edad

# --- Gráfico ---
ax = tabla.plot(kind="bar", stacked=True, figsize=(10,6))

# --- Personalización ---
plt.title("Influencia de la edad en el tipo de institución de ingreso")
plt.ylabel("Porcentaje (%)")
plt.xlabel("Rango de edad")
plt.legend(title="Tipo de institución", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.xticks(rotation=45)
plt.tight_layout()

# --- Mostrar ---
plt.show()


# =====================================================================
# Relacion con otras variables
# =====================================================================

# Seleccionar las 5 instituciones con más estudiantes
top5_instituciones = df['NOMBRE DE INSTITUCION'].value_counts().head(5).index
df_top5 = df[df['NOMBRE DE INSTITUCION'].isin(top5_instituciones)]

# Limitar a las 10 combinaciones Institución + Jornada más frecuentes
top10_combo = (
    df_top5.groupby(['NOMBRE DE INSTITUCION', 'JORNADA'])
    .size()
    .nlargest(10)
    .index
)

df_top10 = df_top5[
    df_top5.set_index(['NOMBRE DE INSTITUCION', 'JORNADA']).index.isin(top10_combo)
]

# Tabla de frecuencias cruzadas: Edad vs Top 10 combinaciones Institución/Jornada
tabla_frec_jornada = pd.crosstab(
    df_top10['RANGO EDAD'],
    [df_top10['NOMBRE DE INSTITUCION'], df_top10['JORNADA']],
    normalize="index"
) * 100

# --- Gráfico ---
tabla_frec_jornada.plot(kind="bar", stacked=True, figsize=(12,6))
plt.title("Influencia de la edad en las 10 principales combinaciones Institución/Jornada")
plt.ylabel("Porcentaje (%)")
plt.xlabel("Rango de edad")
plt.legend(title="Institución / Jornada", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Tabla de frecuencias cruzadas: Edad vs Top 5 Instituciones vs Año de Ingreso
tabla_frec_anio = pd.crosstab(
    df_top5['RANGO EDAD'],
    [df_top5['NOMBRE DE INSTITUCION'], df_top5['AÑO INGRESO']],
    normalize="index"
) * 100
```

```python
# =====================================================================
# Interpretación de medidas de tendencia central.
# =====================================================================

media_por_institucion = df.groupby('NOMBRE CARRERA')['VALOR ARANCEL (PESOS)'].mean()
mediana_por_institucion = df.groupby('NOMBRE CARRERA')['VALOR ARANCEL (PESOS)'].median()

# Corrección de la moda
moda_por_institucion = df.groupby('NOMBRE CARRERA')['VALOR ARANCEL (PESOS)'].apply(
    lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan
)

print("Promedio de arancel por carrera:\n", media_por_institucion)
print("\nMediana de arancel por carrera:\n", mediana_por_institucion)
print("\nModa de arancel por carrera:\n", moda_por_institucion)

# =====================================================================
# Interpretación de medidas de dispersión.
# =====================================================================

maximo_arancel_por_carr = df.groupby('NOMBRE CARRERA')['VALOR ARANCEL (PESOS)'].max()
minimo_arancel_por_carr = df.groupby('NOMBRE CARRERA')['VALOR ARANCEL (PESOS)'].min()
rango_valor = maximo_arancel_por_carr - minimo_arancel_por_carr

print("\nRango de valor por carrera:\n", rango_valor)

desviacion_valor_x_carr = df.groupby('NOMBRE CARRERA')['VALOR ARANCEL (PESOS)'].std()
print("\nDesviación estándar del valor de cada carrera:\n", desviacion_valor_x_carr)

# Coeficiente de variación (en %)
cv_arancel = desviacion_valor_x_carr / media_por_institucion * 100
print("\nCoeficiente de variación del valor de cada carrera (%):\n", cv_arancel)

# =====================================================================
# Análisis con percentiles
# =====================================================================

# Percentiles de los aranceles (global)
percentiles = df['VALOR ARANCEL (PESOS)'].quantile([0.25, 0.5, 0.75, 0.90, 0.95, 1])
print("\nPercentiles de los aranceles:\n", percentiles)

# Definimos umbral "caro" como el percentil 90
umbral_caro = df['VALOR ARANCEL (PESOS)'].quantile(0.90)
print(f"\nUmbral de carreras caras (p90): {umbral_caro}")

# Filtramos las carreras cuyo arancel sea mayor al percentil 90
carreras_caras = df[df['VALOR ARANCEL (PESOS)'] > umbral_caro]
print("\nCarreras sustantivamente más caras:")
print(carreras_caras[['AREA CONOCIMIENTO', 'NOMBRE CARRERA', 'VALOR ARANCEL (PESOS)']])

# =====================================================================
# Relacion con otras variables
# =====================================================================

# Análisis de carreras caras según el nivel de estudio

promedio_por_nivel = df.groupby('NIVEL DE ESTUDIO CARRERA')['VALOR ARANCEL (PESOS)'].mean().sort_values(ascending=False)
print("\nArancel promedio por nivel de estudio:")
print(promedio_por_nivel)

# Gráfico de barras
plt.figure(figsize=(8, 5))
promedio_por_nivel.plot(kind='bar', color='salmon', edgecolor='black')
plt.title("Arancel promedio por nivel de estudio", fontsize=16)
plt.xlabel("Nivel de estudio", fontsize=13)
plt.ylabel("Valor promedio arancel (Pesos)", fontsize=13)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
